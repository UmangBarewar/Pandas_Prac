# Manipulating Data


```python
miss=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-missing-data.csv")
miss
```

```python
miss["Odometer"].fillna(miss["Odometer"].mean(),inplace=True)
miss["Odometer"]
```

```python
seats=pd.Series([5,5,5,5,5])
miss["Seats"]=seats
```

```python
miss
```

```python
miss["Seats"].fillna(5,inplace=True)
```

```python
miss
```

```python
Fuel_eco=[14,7,9,6,22,12,23,19,20,10]
miss["Fuel_eco"]=Fuel_eco
```

```python
miss
```

```python
miss["random"]=miss["Odometer"]*5
```

```python
miss
```

```python
miss["Wheels"]=4
miss
```

```python
miss.sample(frac=0.5)
```

```python
miss["Odometer (miles)"]=miss["Odometer"].apply(lambda x:x/1.6)
```
